#### **Uji Hipotesis**
Apakah Total Frekuensi Berpengaruh Secara Signifikan terhadap User Churned

`H0` = Total Frekuensi Berpengaruh terhadap User Churn 

`H1` = Total Frekuensi tidak berkorelasi terhadap User Churn 





```python
df_place_fav
```

```python
from scipy.stats import chi2_contingency
con_table = pd.crosstab(
    df_place_fav['label'],
    df_place_fav['total_fav']
)


stat,pvalue, dof, exp_val = chi2_contingency(con_table)

alpha = 0.05
if pvalue > alpha:
    print(f'''
    pvalue = {pvalue:.5f}. pvalue > {alpha}. Gagal Menolak H0(Fail To Reject)
    Kita TIDAK Punya Cukup Bukti bahwa antara Label dan Total Frekuensi ada keterkaitan
''')
else:
    print(f'''
    pvalue = {pvalue:.5f}. pvalue <= {alpha}. Menolak H0(Reject H0).
    Kita Punya Cukup Bukti bahwa antara Label dan Total Frekuensi ada keterkaitan
''')
```

```python
churn = df_place_fav[df_place_fav['label'] == 'churned']['total_fav']
retain = df_place_fav[df_place_fav['label'] == 'retained']['total_fav']


from scipy.stats import mannwhitneyu

stat,pvalue = mannwhitneyu(
    x = churn,
    y = retain,
    alternative='two-sided'     #coba perhatikan lagi terkait penggunaan mehthod(two-sided/greater)
)

alpha = 0.05
if pvalue > alpha:
    print(f'''
    pvalue = {pvalue:.5f}. pvalue > {alpha}. Gagal Menolak H0(Fail To Reject)
    Kita TIDAK Punya Cukup Bukti Bahwa Median User Churn TIDAK SAMA DENGAN
    median user Retained  
''')
else:
    print(f'''
    pvalue = {pvalue:.5f}. pvalue <= {alpha}. Menolak H0(Reject H0).
    Kita Punya Cukup Bukti Bahwa Median User Churn TIDAK SAMA DENGAN
    median user Retained  
''')
```

```python
#membuat segmentasi user waze
print(len(df_loyal[df_loyal['n_days_after_onboarding'] <= 365]))
print(len(df_loyal[(df_loyal['n_days_after_onboarding'] > 365) & (df_loyal['n_days_after_onboarding'] <= 1095)]))
print(len(df_loyal[(df_loyal['n_days_after_onboarding'] > 1095) & (df_loyal['n_days_after_onboarding'] <= 2190)]))
print(len(df_loyal[(df_loyal['n_days_after_onboarding'] > 2190) & (df_loyal['n_days_after_onboarding'] <= 2555)]))
```

```python
def segment_days(x):
    if x <= 365:
        return 1
    elif x > 365 and x <= 1095 :
        return 2
    elif x > 1095 and x <= 2190 :
        return 3
    else:
        return 4
    
df_loyal['segmentByAgeInApp'] = df_loyal['n_days_after_onboarding'].apply(segment_days)
df_loyal.sample(5,random_state=12)
```

##### Membuat Kolom Segmentasi Users Berdasarkan driving_days(frekuensi pemakaian dalam hari selama sebulan)
- `0-5 hari kategori 1`
- `6-12 hari kategori 2`
- `13-19 hari kategori 3`
- `>= 20 hari kategori 4`

```python
#membuat segmentasi user berdasarkan frekuensi pemakaian min 1 km dalam hari selama sebulan 

df_loyal['segmentByFreqInMonth'] = pd.qcut(df_loyal['driving_days'], q=4, labels=[1,2,3,4]).astype(int)
df_loyal.sample(5,random_state=42)
```

##### Membuat Kolom Segmentasi Users Berdasarkan duration_minutes_drives(total durasi berkendara dalam menit selama sebulan)
- `18 - 840.18 menit kategori 1`
- `840.18 - 1479.39 menit kategori 2`
- `1479.39 - 2466.92 menit kategori 3`
- `>= 2466.92 hari kategori 4`

```python
#membuat segmentasi user beradarkan total durasi berkendara dalam menit selama sebulan 

df_loyal['segmentByDuration'] = pd.qcut(df_loyal['duration_minutes_drives'],q=4, labels=[1,2,3,4]).astype(int)
df_loyal.sample(5,random_state=42)
```

```python
df_loyal['TotalSegment'] = df_loyal['segmentByAgeInApp'] + df_loyal['segmentByFreqInMonth'] + df_loyal['segmentByDuration']
df_loyal.sample(5,random_state=42)
```

```python
#membuat category level berdasarkan segmentation 
def total_segmen(x):
    if x <= 5:
        return 'Low'
    elif x <= 7:
        return 'Medium'
    elif x <= 10:
        return 'Premium'
    else: 
        return 'Royal'
    
df_loyal['lvl_user'] = df_loyal['TotalSegment'].apply(total_segmen)
df_loyal.sample(5,random_state=42)
```
